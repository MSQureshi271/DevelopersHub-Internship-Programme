# Task 04: General Health Query Chatbot 

## Import Libraries

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
```

## Load the .env 

```python
load_dotenv()
client = genai.Client(api_key=os.getenv("GENAI_API_KEY"))
```

## Provide a system prompt

```python
system_prompt = """
You are a friendly medical information assistant.
Give general health information only.
Do not diagnose.
Do not prescribe medicine or dosage.
For serious symptoms, tell the user to seek urgent medical help.
Always remind the user that this is general information, not medical advice.
"""
```

## Chatbot function

```python
def health_chatbot(question):
    emergency_words = [
        "chest pain",
        "difficulty breathing",
        "severe bleeding",
        "unconscious",
        "stroke",
        "seizure",
        "suicide",
        "overdose"
    ]

    lower_question = question.lower()

    for word in emergency_words:
        if word in lower_question:
            return "Please seek emergency medical help immediately. This is general information, not a medical diagnosis."

    prompt = f"""
{system_prompt}

User question:
{question}

Answer clearly and safely:
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0.4,
        )
    )

    return response.text
```

## Test the Health Chatbot

```python
print(health_chatbot("What causes a sore throat?"))
```

```python
print(health_chatbot("Is paracetamol safe for children?"))
```
