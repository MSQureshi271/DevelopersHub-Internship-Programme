# Task 1: Exploring and Visualizing a Simple Dataset

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset using Pandas

```python
df = pd.read_csv('iris.csv')
df.head()
```

## Summary Statistics with .info() and .describe()

```python
df.info()
```

```python
df.describe()
```

## Visualize the provided Data

### a) Scatter Plot

#### Matplotlib

```python
plt.scatter(df['sepal.length'], df['sepal.width'], c='red')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.scatter(df['petal.length'], df['petal.width'], c='blue')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Dataset: Petal Length vs Petal Width')
plt.show()
```

#### Seaborn

```python
sns.scatterplot(data = df, x='sepal.length', y='sepal.width', hue='variety')
```

```python
sns.scatterplot(data=df, x='petal.length', y='petal.width', hue='variety')
```

### b) Histograms

#### Matplotlib

```python
features = df.columns[:-1]

for col in features:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20, color=(col=='sepal.length' or col=='sepal.width') and 'red' or 'blue', edgecolor='black')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col}')
    plt.show()
```

#### Seaborn

```python
features = df.columns[:-1]

for col in features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=20, color=(col=='sepal.length' or col=='sepal.width') and 'red' or 'blue')
    plt.title(f"Distribution of {col}")
    plt.show()
```

### c) Box Plot

#### Matplotlib

```python
numeric_cols = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

plt.figure(figsize=(8, 5))
plt.boxplot([df[col] for col in numeric_cols], labels=numeric_cols)
plt.title('Boxplot of Iris Dataset Features')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Seaborn

```python
df_melted = df.melt(
    id_vars='variety',
    value_vars=numeric_cols,
    var_name='Feature',
    value_name='Value'
)

plt.figure(figsize=(12,6))

sns.boxplot(
    data=df_melted,
    x='Feature',
    y='Value',
    hue='variety'
)

plt.title('Feature Distribution by Species')
plt.xticks(rotation=45)

plt.show()
```
